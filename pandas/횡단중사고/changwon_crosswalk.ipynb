```python
import csv, re
import pandas as pd

re_word = re.compile(r'창원시 (.*구) ')

data = {
    '지역': [],
    '2012': [],
    '2013': [],
    '2014': [],
    '2015': [],
    '2016': [],
    '2017': [],
    '2018': [],
    '2019': []
}

with open('changwon_crosswalk_modded.csv', encoding='utf-8-sig') as f:
    cells = csv.DictReader(f)

    for k in cells:
        index = -1
        region = re_word.search(k['주소']).group(1)

        if region in data['지역']:
            index = data['지역'].index(region)
        else:
            data['지역'].append(region)
            data['2012'].append(0)
            data['2013'].append(0)
            data['2014'].append(0)
            data['2015'].append(0)
            data['2016'].append(0)
            data['2017'].append(0)
            data['2018'].append(0)
            data['2019'].append(0)

        for i in range(2, 10):
            if '201%s' % i in k['발생년']:
                data['201%s' % i][index] += 1

result = pd.DataFrame(data)
result

```
